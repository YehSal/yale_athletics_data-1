```python
# Clean up High Schools data and make it consistent with players data
import os
import pandas as pd
import matplotlib.pyplot as plt
PROJ_ROOT = os.pardir

# For munging
import re
import json
```

```python
playersPath = os.path.join(PROJ_ROOT, "data", "processed", "player_bios_processed.csv")
players = pd.read_csv(playersPath)
```

```python
players.shape
```

```python
players.head()
```

```python
schools = players["High School"].unique()
ids = range(0, len(schools))
```

```python
highSchools = pd.DataFrame({
        "High_School_ID": ids,
        "High School": schools
    })
```

```python
highSchools = highSchools[["High_School_ID", "High School"]]
highSchools.set_index("High_School_ID")
```

```python
# Seems about right.
processedPath = os.path.join(PROJ_ROOT, "data", "processed", "high_schools_processed.csv")
highSchools.to_csv(processedPath, encoding='utf-8')
```
